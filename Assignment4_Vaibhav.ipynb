```python
import pandas as pd
```

```python
df=pd.DataFrame()
print(df)
```

```python
df=pd.read_csv(r'D:\Cleaned_data.csv')
df
```

```python
# 1. Display all unique countries in the dataset.
unq =df['country'].unique()
print(unq)
```

```python
#2. Find the total number of years available.
years=df['year'].nunique()
print(years)
```

```python
#3. What is the earliest and latest year in the dataset? 
earliest = df['year'].min()
latest= df['year'].max()
print('earliest='+str(earliest))
print('latest='+str(latest))
```

```python
#4. What is the total CO₂ emission in the most recent year?
total=df[df['year']==latest]['co2'].sum()
print(total)
```

```python
#5. Find the top 5 countries with the highest total CO₂ emissions.
top_5= df.groupby('country')['co2'].sum().sort_values(ascending=False)
top_5.head()
```

```python
#6. Find the average CO₂ emission per year for India.
avg_co2 = df[df['country'] == 'India']['co2'].mean()
print(avg_co2)
```

```python
#7. What is the maximum and minimum CO₂ emission recorded? 
co2_min=df['co2'].min()
co2_max=df['co2'].max()
print('minimum='+str(co2_min))
print('maximum='+str(co2_max))
```

```python
#8. Which country had the lowest CO₂ emissions in 2010?
df_2010 = df[df['year'] == 2010]
lowest = df_2010.loc[df_2010['co2'].idxmin(), 'country']
lowest
```

```python
#9. Group the data by country and find the average CO₂ emission per country.
avg_per_country = df.groupby('country')['co2'].mean()
print(avg_per_country)
```

```python
#10. Sort the dataset by CO₂ emission in descending order.
sort = df.sort_values(by='co2', ascending=False)
sort
```
